```python
import pandas as pd
import math
```

```python
# Load into dataframe
dataframe = pd.read_csv("nyc_housing_base.csv")
```

```python
# Remove duplicate data
dataframe = dataframe.drop_duplicates()
```

```python
# Drop borough_x
dataframe = dataframe.drop("borough_x", axis=1)
```

```python
# Look for correlations within entire dataset
all_corr = dataframe.corr(method='spearman', numeric_only=True)
```

No strong correlations were found between numerical values and the sale price for the entire dataframe. Let's examine relationships for each borough.

```python
# Get available borough names
boroughs = dataframe["borough_y"]
boroughs = boroughs.drop_duplicates()
borough_names = boroughs.values
borough_dict = {}

for borough in borough_names:
    borough_df = dataframe.query(f"borough_y == '{borough}'")
    borough_dict[borough] = borough_df.corr(method='spearman', numeric_only=True)
```

```python
print(borough_dict[borough_names[4]])
```

No significant correlations were seen between raw variables and sale price. Let's analyze trends in residential density vs. sale price.

```python
# Residential density = unitsres / lotarea
dataframe["resdens"] = dataframe["unitsres"] / dataframe["lotarea"]
```

```python
# Get available zip codes
zips = dataframe["zip_code"].dropna()
zips = zips.drop_duplicates()
zip_vals = zips.values
zip_dict = {}

for zc in zip_vals:
    zip_df = dataframe.query(f"zip_code == {zc}")
    zip_dict[zc] = zip_df.corr(method='spearman', numeric_only=True)
```

```python
print(zip_dict[zip_vals[0]]["resdens"])
```
